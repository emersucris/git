# Importando Bibliotecas


```python
# Importando Bibiliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importando Dados

```python
# Criando variavel df
df = pd.read_csv("/content/dados_rh.csv")
```

```python
# Exibindo os dados da variavel
df
```

```python
# Preparando o dataset
departments = df['Department'].unique()
employee_counts = df.groupby('Department')['EmployeeCount'].sum()
```

```python
# Criando o bar plot
plt.bar(departments, employee_counts)

# Adicionando aslabels e o titulo
plt.xlabel('Departmento')
plt.ylabel('Número de empregados')
plt.title('Empregados por departamento')
```

```python
# Agrupando os dados do dataframe
education_counts = df.groupby("Education").size()
```

```python
# Plotando o gráfico
plt.plot(education_counts.index, education_counts.values)
plt.title("Número de funcionários por nível de escolaridade")
plt.xlabel("Nivel de escolaridade")
plt.ylabel("Número de empregados")
```

```python
# Colocando dados do dayaframe
labels = ['Não', 'Sim']
sizes = [df['Attrition'].value_counts()[0], df['Attrition'].value_counts()[1]]

# Criando um gráfico de pizza
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures a circular pie chart

# Adicionando titulo
ax.set_title('Quantidade de funcionários com e sem atrito')
```
