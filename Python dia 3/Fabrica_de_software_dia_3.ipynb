# Mostrando a média em matemática para cada grupo disponivel

```python
# Agrupando e plotando o gráfico
df2 = df.groupby("race/ethnicity")["math score"].mean().reset_index()
plt.bar(df2["race/ethnicity"], df2["math score"])
plt.title("Average Math Scores by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Average Math Score")
plt.show()
```

# Media de cada grupo a partir do nivel de escolaridade dos parentes

```python
# Agrupando e plotando o gráfico
grouped_df = df.groupby("parental level of education")["math score"].mean().reset_index()
plt.bar(grouped_df["parental level of education"], grouped_df["math score"])
plt.title("Average Math Scores by Parental Level of Education")
plt.xlabel("Parental Level of Education")
plt.ylabel("Average Math Score")
```

# Agrupando o banco de dados a partir da coluna de participação em um curso preparatorio para testes

```python
# Criando a variavel
df3 = df.groupby("test preparation course")["reading score"].mean().reset_index()
```

```python
# Plotando o gráfico
plt.bar(df3["test preparation course"], df3["reading score"])
plt.title("Average Reading Scores by Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Reading Score")
plt.show()
```
